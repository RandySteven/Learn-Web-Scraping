```python
import requests
from bs4 import BeautifulSoup
import csv

page = requests.get('https://codedamn-classrooms.github.io/webscraper-python-codedamn-classroom-website/')
soup = BeautifulSoup(page.content, 'html.parser')

all_products = []

products = soup.select('div.thumbnail')

for product in products :
  name = product.select('h4 > a')[0].text.strip()
  description = product.select('p.description')[0].text.strip()
  price = product.select('h4.price')[0].text.strip()
  reviews = product.select('div.ratings')[0].text.strip()
  image = product.select('img')[0].get('src')

  all_products.append({
        "name": name,
        "description": description,
        "price": price,
        "reviews": reviews,
        "image": image
  })

for product in all_products:
  print(product)
```

```python
from enum import Enum

class Job:

  def __init__(self, jobTitle, company, startDuration, EndDuration, companyLocation, jobDescription, keySkills):
    self.jobTitle = jobTitle
    self.company = company
    self.startDuration = startDuration
    self.endDuration = endDuration
    self.companyLocation = companyLocation
    self.jobDescription = jobDescription
    self.keySkills = keySkills
```

```python
from enum import Enum
class Time(Enum):
  START = 0
  END = 1 
```

```python
import json
import csv
import pandas as pd

all_jobs = []

def contractDurationSpliter(contractDuration, index):
  indexPlus = 0
  contractDurationTxt = contractDuration
  indexPlus = index + 1
  time = contractDurationTxt.split('-', indexPlus)
  return time[index]

def keySkillsSpliter(keySkills):
  keySkillsTxt = keySkills
  keySkillArr = keySkillsTxt.split(' , ')
  return keySkillArr

for job in jobs:
  jobTitle = job.select('h2 > a')[0].text.strip()
  company = job.select('h3.joblist-comp-name')[0].text.strip().replace("\\\\n", '')
  contractDuration = job.select('ul > li')[0].text.strip().replace('card_travel', '').replace('yrs', '').replace(' ', '')
  startDuration = contractDurationSpliter(contractDuration, Time.START.value)
  endDuration = contractDurationSpliter(contractDuration, Time.END.value)
  companyLocation = job.select('ul > li')[1].text.strip().replace('location_on', '')
  jobDescription = job.select('ul.list-job-dtl.clearfix > li')[0].text.strip().replace('Job Description', '').replace(":\\\\r\\\\n", '')
  keySkills = keySkillsSpliter(job.select('ul.list-job-dtl.clearfix > li')[1].text.strip().replace('KeySkills', '').replace(':\n\r\n      \r\n          ', ''))
  

  objJob = Job(jobTitle, company, startDuration, endDuration, companyLocation, jobDescription, keySkills)

  all_jobs.append(objJob)
```

```python
from enum import Enum
class FileFormat(Enum):
  CSV = 1
  JSON = 2
```

```python
class FileUtil:

  def FileWriter(fileName, array, fileFormat):
    if(fileFormat == 2):
      with open(fileName, 'w') as f:
        f.write(str(array).replace("'", ""))
    elif(fileFormat == 1):
      with open(fileName, 'w') as f:
        f.write(str(array).replace("'", ""))
```

```python
jobs_json = []

for job in all_jobs:
  jobs_json.append(json.dumps(vars(job)))

str(jobs_json).replace("'", "")

FileUtil.FileWriter("jobs.json", jobs_json, FileFormat.JSON.value)
FileUtil.FileWriter("jobs.csv", jobs_json, FileFormat.CSV.value)
```
